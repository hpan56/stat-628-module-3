```python
import pandas as pd
import json_to_df as jd
import spacy
import json
```

```python
## initialize the package
nlp = spacy.load('en_core_web_sm')
nlp.Defaults.stop_words -= {"no", "n't", "never", "neither", "nor", "nobody", "none", "nothing", "noone", "not", "nowhere",
                           "n’t", "n‘t"}
```

```python
## Functions to discover the structure.
def text_convert(text, dicts, nlp):
    review_ls = []
    doc = nlp(text)
    
    for i in doc.sents:
        ## Collect heads into a dict
        heads = {}
        ## Initial a root key for the dict
        heads["root"] = [i.root.lemma_]
        sent = i.as_doc()
        for j in sent:
            if ((j.is_punct) | (j.text in nlp.Defaults.stop_words) | (j.pos_ == "SPACE") | (j.lemma_ == "-PRON-")):
                continue
            if j.lemma_ not in dicts.keys():
                dicts[j.lemma_] = ([1], j.pos_)
            else:
                dicts[j.lemma_][0][0] += 1
            if j.head.lemma_ not in heads.keys():
                heads[j.head.lemma_] = [j.lemma_]
            else:
                heads[j.head.lemma_].append(j.lemma_)
        ## Supply the words for root-word
        new_ls = []
        for j in sent:
            if j.lemma_ in heads.keys():
                try:
                    if j.lemma_ not in heads[j.head.lemma_]:
                        heads[j.head.lemma_].append(j.lemma_)
                except:
                    try:
                        if j.head.lemma_ not in heads.keys():
                            heads[j.head.lemma_] = [j.lemma_]
                        else:
                            heads[j.head.lemma_].append(j.lemma_)
                        new_ls.append(j.head.lemma_)
                    except:
                        pass
        for j in sent:
            if j.lemma_ in new_ls:
                try:
                    heads[j.head.lemma_].append(j.lemma_)
                except:
                    pass
        ## add the dict to a list
        review_ls.append(heads)
            
    return review_ls, dicts
```

```python
## initialize a word table
tables = {}
## Read Data
text = jd.json_to_df_lines('../raw_data/review_bus_user.json',100)
for i in range(len(text)):
    text.text[i], tables = text_convert(text.text[i], tables, nlp)
```
