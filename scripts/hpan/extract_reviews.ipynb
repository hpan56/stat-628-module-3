```python
import pandas as pd
import json
import json_to_df as jd
import date_converting as dc
```

```python
reviews = jd.json_to_df_lines(path = "../raw_data/review.json", lines = 1600000, skip = 3200000)
```

```python
reviews.head()
```

```python
business = jd.json_to_df_lines(path = "../raw_data/business.json", lines = 192609)
business.rename(columns={"stars":"bus_stars", "review_count":"bus_review_count",}, inplace = True)
```

```python
business_id = pd.read_table("../raw_data/gyms.txt", header = None)
business_id.rename(columns={0:"business_id"}, inplace = True)
business_id = business_id.merge(business, how = "left", on = "business_id")
```

```python
business_id.head()
```

```python
reviews_bus = reviews.merge(business_id, how = "inner", on = "business_id")
reviews_bus.shape
```

```python
users = jd.json_to_df_lines(path = "../raw_data/user.json", lines = 1637138)
```

```python
users.rename(columns={"name":"user_name", "review_count":"user_review_count", "useful":"user_useful",
                      "funny":"user_funny", "cool":"user_cool"}, inplace=True)
```

```python
reviews_bus_user = reviews_bus.merge(users, how = "inner", on = "user_id")
reviews_bus_user.head()
```

```python
reviews_bus_user = reviews_bus_user.drop(columns=["review_id", "address", "postal_code", "latitude", 
                                                  "longitude", "attributes", "categories", "hours", "user_name"])
```

```python
reviews_bus_user.date = reviews_bus_user.date.apply(dc.get_days)
reviews_bus_user.yelping_since = reviews_bus_user.yelping_since.apply(dc.get_days)
reviews_bus_user.friends = reviews_bus_user.friends.apply(lambda x: len(x.split(","))) 
```

```python
a = reviews_bus_user.to_json(orient = 'records').replace('},{', '}\n{').replace("[","").replace("]","")
```

```python
with open('../raw_data/review_bus_user.json', 'w') as f:
    f.write(a)
```
